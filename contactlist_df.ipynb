VISUALIZATION

```python
#total number of males and females on the contact list
import matplotlib.pyplot as plt
import seaborn as sbn

sbn.set_style('darkgrid')
contacts_df.Gender.value_counts().plot(kind='bar')
plt.title('Gender Distribution of Contact List')
plt.ylabel('Frequency')
plt.xticks(rotation=75)
```

```python
#persons met
import matplotlib.pyplot as plt
import seaborn as sbn

sbn.set_style('white')
plt.figure(figsize=(5,5))
#contacts_df.Met.value_counts().plot(kind='bar', color = ['g', 'r'])
contacts_df.Met.value_counts().plot(kind='pie')
plt.legend()
plt.xticks(rotation=70)
plt.xlabel('Persons Met')
plt.title('Persons Met Distribution')
```

```python
#persons met by gender
import matplotlib.pyplot as plt
import seaborn as sbn

sbn.set_style('darkgrid')
plt.figure(figsize=(5,4))
#contacts_df[['Gender', 'Met']].value_counts().sort_index(ascending=True).plot(kind='barh', color=['r','g'])
contacts_df[['Gender', 'Met']].value_counts().sort_index(ascending=False).plot(kind='bar', color=['g','r'])
plt.xticks(rotation=70)
plt.title('Persons Met by Gender')
plt.ylabel('Count')

```

```python
#the top network used
import matplotlib.pyplot as plt
import seaborn as sbn


sbn.set_style('darkgrid')
#plt.figure(figsize=(6,4))
#plt.plot(contacts_df.Network.value_counts(), marker='*', color='r')
contacts_df.Network.value_counts().sort_values(ascending=True).plot(kind='barh', color=(['g','b','r','y']))
#plt.xticks(rotation=75)
#plt.yticks(rotation=70)
plt.xlabel('Frequency')
plt.ylabel('Network')
plt.title('Network Distribuution of Contacts')
```

```python
import matplotlib.pyplot as plt
import seaborn as sbn

sbn.set_style('dark')
contacts_df.Met.value_counts().plot(kind='pie')
plt.legend()
plt.title('Persons Met Distribution')

```

```python
import matplotlib.pyplot as plt
import seaborn as sbn

sbn.set_style('darkgrid')
#plt.bar(contacts_df[['Network', 'Gender']].value_counts().sort_index())
#contacts_df[['Network', 'Gender']].value_counts().sort_index().plot(kind='bar')
contacts_df[['Network', 'Gender']].value_counts().sort_index().plot(kind='bar', color=['b','b','y','y','g','g','r','r'])

plt.xticks(rotation=80)
plt.ylabel('Frequency')
plt.title('Network Distribution by Gender')
#plt.bar_label(label_type=Literal['center'])

```
